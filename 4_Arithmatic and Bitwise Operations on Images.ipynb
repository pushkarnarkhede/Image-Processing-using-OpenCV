### Arithmatic operations on Images
- These are simple operation which allows you to directly add or subtract to image pixels.
- these operations are generally used when you want to deal with pixels values or intensity

```python
# define a function which display the image read by cv2 using matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
import numpy as np

def show_image(image = None, title = "", size = 10):
    if image is not None:
        height, width = image.shape[0], image.shape[1]
        aspect_ratio = width / height
        
        plt.figure(figsize=(size * aspect_ratio, size))
        rgb_image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
        
        plt.imshow(rgb_image)
        plt.title(title)
        plt.show()
    else:
        print("Image is null")
```

```python
# Let's brighten the image using cv2.add(img,value_or_matrix_to_add) 
path = "images/132.jpg"
img = cv2.imread(path)
show_image(img, "Original Image", 3)

# in order to bright the image we have to add some value to each original pixels of an image
bright_img = cv2.add(img,100) # here i am adding the 100 to each pixels so that the image looks more brighter 
show_image(bright_img, "Bright Image using Single value", 3)

# the another approach is to create the matrix of same size of an image and perform matrix addition using cv2.add() method
M = np.ones(img.shape,np.uint8) * 100 # here i am creating the matrix of shape 150x150x3 and multiplying each value with 100 so that the matrix constain 100 value
bright_img2 = cv2.add(img,M)
show_image(bright_img2, "Bright Image using Matrix", 3)
```

```python
# Let's dark the same image using cv2.subtract(img, value_or_matrix_to_add) 
path = "images/132.jpg"
img = cv2.imread(path)
show_image(img, "Original Image", 3)

# in order to bright the image we have to subtract some value to each original pixels of an image
bright_img = cv2.subtract(img,100) # here i am subtract the 100 to each pixels so that the image looks more brighter 
show_image(bright_img, "Dark Image using Single value", 3)

# the another approach is to create the matrix of same size of an image and perform matrix subtraction using cv2.subtract() method
M = np.ones(img.shape,np.uint8) * 100 # here i am creating the matrix of shape 150x150x3 and multiplying each value with 100 so that the matrix constain 100 value
bright_img2 = cv2.subtract(img,M)
show_image(bright_img2, "Dark Image using Matrix", 3)
```

### Bitwise Operations on images
- Bitwise operation usef for combining or merging two images.

```python
# Let's use and, or, xor, not bitwise operation to understand how they works
# let's first create a square image and draw a square into it
square = np.zeros((300,300), np.uint8)
cv2.rectangle(square, (50,50), (200,200), (255,0,0), -1)

# let's create second image of square with circle in it
circle = np.zeros((300,300), np.uint8)
cv2.circle(circle, (175,175), 100, (255,0,0), -1)

show_image(square, "Square inside Square image", 3)
show_image(circle, "Circle inside Square image", 3)
```

```python
# use cv2.bitwise_and(img1, img2) to And both image
bitwise_and_img = cv2.bitwise_and(square,circle)
show_image(bitwise_and_img, "Bitwise And image", 3)

# use cv2.bitwise_or(img1, img2) to combine bith images
bitwise_or_img = cv2.bitwise_or(square,circle)
show_image(bitwise_or_img, "Bitwise Or image", 3)

# use cv2.bitwise_xor(img1, img2) to remove the common color portion and add both images
bitwise_xor_img = cv2.bitwise_xor(square,circle)
show_image(bitwise_xor_img, "Bitwise Xor image", 3)

# use cv2.bitwise_not(img1)
bitwise_not_img = cv2.bitwise_not(square)
show_image(bitwise_not_img, "Bitwise Not image", 3)
```
