```python
# import the opencv module
import cv2 
print(cv2.__version__) # let's check installd version of opencv module
```

```python
# To load an image we have to follow 2 basics steps
# 1. we have to first set the path of the image to load
# 2. use cv2.imread(path) method to load the image as an n x m x c array

# let's set the Path of image
IMG_PATH = "images/119.jpg"

# use cv2.imread(path) method to load image
img = cv2.imread(IMG_PATH)

# print the "img" variable to see what opencv has loaded
print(img)
```

```python
# Let's first understand the shape of image and what type of image it is, Color or Grayscale
# opencv has .shape method which return the shape in the form of (height, width, channel)

image_shape = img.shape

# the loaded image is of shape 150 pixels x 150 pixels with 3 channel
# the 3 channel means the image as 3 2D array of size 150 x 150 
# in our case the array are RGB i.e Red, Green and Blue which makes it color image
# when the channel value is 1 it means the image is grayscale and it have only 1 2D array.
# there are many images with 4 channels where the 4th channel represent another 2D array.
# in case of RGBA the 4th channel is Alpha which represent the transparancy of image.

print("The shape of image is => ", image_shape) 
print("Height of image is => " ,image_shape[0])
print("Width of image is => " ,image_shape[1])
print("Number of channel in the image is => " ,image_shape[2])
```

```python
# let's display the loaded image into the jupyter notebook
# use matplotlib module to display the loaded image

# using matplotlib 
import matplotlib.pyplot as plt
%matplotlib inline 

# as we have loaded the image using opencv the image is in the BGR format and hence to display image using plt
# we have to convert the image from BGR to RGB for that use cv2.cvtColor(img,cv2.COLOR_BGR2RGB) 
# cv2.cvtColor() method used to convert the image from any color channel to any color channel
# for us i am specifying cv2.COLOR_BGR2RGB which convert the img from BGR to RGB

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
# use cv2.imwrite(path_to_save_image.jpg/png,img) to save any loaded image
SAVE_PATH = "images/mountainImage.jpg"
cv2.imwrite(SAVE_PATH, img)
```

```python
# define a function which display the image read by cv2 using matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import cv2

def show_image(image = None, title = "", size = 10):
    if image is not None:
        height, width = image.shape[0], image.shape[1]
        aspect_ratio = width / height
        
        plt.figure(figsize=(size * aspect_ratio, size))
        rgb_image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
        
        plt.imshow(rgb_image)
        plt.title(title)
        plt.show()
    else:
        print("Image is null")
```

```python
# use cvtColor(img,cv2.COLOR_BGR2GRAY) to convet the color image into grayscale
img = cv2.imread(IMG_PATH)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
show_image(image = img, title = "Mountain", size = 5)
```
