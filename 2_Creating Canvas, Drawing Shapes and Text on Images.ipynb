### Creating Canvas

```python
# define a function which display the image read by cv2 using matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
import numpy as np

def show_image(image = None, title = "", size = 10):
    if image is not None:
        height, width = image.shape[0], image.shape[1]
        aspect_ratio = width / height
        
        plt.figure(figsize=(size * aspect_ratio, size))
        rgb_image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
        
        plt.imshow(rgb_image)
        plt.title(title)
        plt.show()
    else:
        print("Image is null")
```

```python
# let's first create a full black image 

# to create a black image/canvas we have to use 0's and for white we have to use 1's 
black_image = np.zeros((512,512,3),np.uint8) # to create rgb black color immage we have to specify the channel i.e 3

grayscale_black_image = np.zeros((512,512),np.uint8) # to create grayscale image we don't neet to specify any channel

white_image = np.ones((512,512,3), np.uint8) * 255 # 255 pixel value represent the white color hence i am multiplying the evey 1 in and image array with 255 to create a image having white pixels

show_image(black_image, "RGB Black Image", 5)
show_image(grayscale_black_image, "GrayScale Black Image", 5)
show_image(white_image, "RGB White Image",5)
```

### Drawing shapes on image and canvas

```python
# use cv2.line(image,(start_x,start_y),(end_x,end_y), (color_B,color_G,color_R), thickness) method to draw a line on any image
# let's draw a line on the newely created RGB Black image

black_image = np.zeros((512,512,3),np.uint8)

cv2.line(black_image,(100,100),(400,400), (0,0,255),2)

show_image(black_image,"Line drawn on black image",5)

# always make sure if you want to change the coordinates of line make sure you have recreated the black image 
# or reload the img using cv2.imread(path) as these function changes the original image it doesn't return anything
```

```python
# use cv2.rectangle(image,(start_x,start_y),(end_x,end_y), (color_B,color_G,color_R), thickness) method to draw rectangle 

black_image = np.zeros((512,512,3),np.uint8)

cv2.rectangle(black_image,(150,150),(350,350), (0,0,255), -1) # if thicknees is -ve then the rectangle will be filled, use positive number if you want unfilled rectangle
#cv2.rectangle(black_image,(150,150),(350,350), (0,0,255), 2)

show_image(black_image,"Rectangle drawn on black image",5)
```

```python
# use cv2.circle(image, (center_X, center_Y), radius, thickness [if +ve then works as thickness, if -ve then works as color fill])
# to draw a circle on any image

black_image = np.zeros((512,512,3),np.uint8)

cv2.circle(black_image,(250,250),200, (0,0,255),5)
cv2.circle(black_image,(250,250),100, (0,255,0),-1)

show_image(black_image,"Filled and unfilled circle drawn on black image",5)
```

```python
# use cv2.polylines(image, array_of_points, is_close_last_and_first_points, color, thickness)
black_image = np.zeros((512,512,3),np.uint8)

# let's define the each point location in an array 
points = np.array( [ [25,150], [350,75], [110,450] ] ,np.int32)
print(points , "\n" , points.shape , "\n")

# reshape the array to 2D plane
points = points.reshape((-1,1,2))
print(points ,"\n", points.shape ) # it represent i have 3 1D array of coordinate x,y

cv2.polylines(black_image, [points], True, (0,0,255), 5)

show_image(black_image,"PolyLines drawn on black image",5)
```

```python
# use cv2.putText(image, text_to_display_on_image,(start_X,start_y), font, fontsize, color, thickness)
black_image = np.zeros((512,512,3),np.uint8)

cv2.putText(black_image, "Hello World", (50,250), cv2.FONT_HERSHEY_COMPLEX_SMALL, 3, (40,200,0), 2)

show_image(black_image,"Text added on black image",5)
```

### Adding a square on image

```python
# let's add a square on bike rider so that it can looks like the bike rider is seleted
path = "images/3.jpg"
img = cv2.imread(path)

show_image(img, "Original Image", 5)

cv2.rectangle(img, (20,65), (50,115), (0,0,255), 1) # the x,y points are calculated from the x,y plane 
show_image(img, "Rectangle created for 1 Bike Rider", 5)

cv2.rectangle(img, (60,60), (85,105), (0,255,0), 1)
show_image(img, "Rectangle created Bike Rider 1 and 2", 5)
```
